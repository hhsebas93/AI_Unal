```python
import pandas as pd
```

```python
class Iris_3:
        def __init__(self, datos):
            self.datos = pd.read_csv(datos)
            self.data = pd.DataFrame(self.datos)
            self.index = 0
        def __iter__(self):
            return self
        def __next__(self):
            if self.index >= len(self.data)-1:
                print('Error')
                raise StopIteration()
            if self.index < len(self.data):
                x= self.data.iloc[self.index], self.data.iloc[self.index+1]
                self.index += 2
                return x
    
```
